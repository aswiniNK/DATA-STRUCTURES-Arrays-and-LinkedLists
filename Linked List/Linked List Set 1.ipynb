## Delete a Node in Linked List

#####  To delete a node from the linked list we need to do three steps, :
    1). Find previous node of the node to be deleted
    2). Change the next of previous node
    3). Free Memory for the node to be deleted



```python
## Python program to delete a node from linked list

# Node class
class Node:
    
    # Constructor to initialize the node object
    def __init__(self, data):
        self.data = data
        self.next = None
        
## Class for LinkedList
class LinkedList4:
    
    # Function to initialize head
    def __init__(self):
        self.head = None
        
    # Function to insert a new node at the beginning
    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node
        
    ## Given the reference to the head of the list and a key, 
    ## delete the first occurence of the key in linked list
    
    def deleteNode(self, key):
        
        # Store head node
        temp = self.head
        
        # If head node itself holds the key to be deleted
        
        if(temp is not None):
            if(temp.data == key):
                self.head = temp.next
                temp = None
                return
            
        ## Search for the key to be deleted, keep track of the 
        ## previous node as we need to change 'prev_next'
        
        while(temp is not None):
            if temp.data == key:
                break
            prev = temp
            temp = temp.next
            
        ## if key was not present in Linked List
        if(temp == None):
            return
        
        # Unlink the node from linked list
        prev.next = temp.next
        
        temp = None
        
    ## Utility function to print the Linked list
    def printList(self):
        
        temp = self.head
        while(temp):
            print(temp.data)
            temp = temp.next
            
## Driver Program

llist = LinkedList4()

llist.push(7)
llist.push(1)
llist.push(3)
llist.push(2)

print("Created Linked List is ")
llist.printList()

llist.deleteNode(1)

print("After Deletion")
llist.printList()
    
        
```

## Deleting a Key at Given Position

```python
## Node class
class Node:
    
    # COnstructor to initialize the object
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList5:
    
    # Constructor to initialize head
    def __init__(self):
        self.head = None
        
    # Function to insert a new node at the beginning
    def push(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node
        
    # Given a reference to the head of the list
    # AND A POSITION, delete the node at a given position
    
    def deleteNode(self, position):
        
        # If linked list is empty
        if self.head == None:
            return
        
        # Store Head Node
        temp = self.head
        
        # if head needs to be removed
        if position == 0:
            self.head = temp.next
            temp = None
            return
        
        # Find previous node of the node to be deleted
        
        for i in range(position-1):
            temp = temp.next
            if temp is None:
                break
                
        # If position is more than number of nodes
        if temp is None:
            return
        if temp.next is None:
            return
        
        # Node temp.next is the node to be deleted 
        # store pointer to the next of node to be deleted
        
        next = temp.next.next
        
        # Unlink the node from linked list 
        temp.next = None
        
        temp.next = next
        
        
    ## utility function to print the linked list
    def printList(self):
        temp = self.head
        while(temp):
            print(temp.data)
            temp = temp.next
            
## Driver function
if __name__ == "__main__":
    
    llist = LinkedList5()
    
    llist.push(7)
    llist.push(1)
    llist.push(3)
    llist.push(2)
    llist.push(8)
    
    print("Create a Linked List")
    llist.printList()
    
    llist.deleteNode(4)
    
    print("Linked List After Deletion at Position 4")
    llist.printList()
        
        
        
```

## Finding Length of the Linked List ( Iterative Approach )

```python
class Node:
    def __init__(self, data):
        self.data = data # Assign data
        self.next = None # Initialize next as null
        
class LinkedList6:
    
    def __init__(self):
        self.head = None
        
    # function is in Linked List 
    # It inserts a new node at beginning of Linked List
    
    def push(self, new_data):
        
        # 1 & 2. Allocate the Node and 
                 # put in the data
        new_node = Node(new_data)
        
        # 3. Make next of new Node as head
        new_node.next = self.head
        
        # 4. Move the head to point to new Node
        self.head = new_node
        
# The function to count number of nodes in the linked list
# iteratively given node as starting node
    def getCount(self):
        temp = self.head # Initializing temp
        count = 0 # Initialize count
    
        # Loop while end of Linked List is not reached
    
        while(temp):
            count += 1
            temp = temp.next
        return count


# Main 

if __name__ == "__main__":
    
    llist = LinkedList6()
    llist.push(1)
    llist.push(3)
    llist.push(1)
    llist.push(2)
    llist.push(1)
    
    print("Count of node is : ", llist.getCount())
    
        
        
```

## Search an element in Linked List (Iterative Approach)

```python
class Node:
    
    def __init__(self, data):
        self.data  = data
        self.next  = None
        
class LinkedList8:
    
    def __init__(self):
        self.head = None  # Initializing head as None
        
    # Function to insert elements into Linked List
    def push(self, new_data):
        
        # 1. Allocate Node
        # 2. Put into Node
        new_node = Node(new_data)
        
        # 3. Making next of new node as head # At first insertion it will be none 
        # later on old nodes are pointed by new nodes 
        new_node.next = self.head
        
        # 4. making the head to point to new node
        self.head = new_node
        
    # Checking wheather the element is present in Linked List    
    def search(self, x):
        
        # Initialize current to head
        current = self.head
        
        # Loop till current not equal to None
        while(current != None):
            if current.data == x:
                return True
            
            current = current.next
            
        return False
    
# Main Method

if __name__ == "__main__":
    
    llist = LinkedList8()
    
    llist.push(7)
    llist.push(13)
    llist.push(27)
    llist.push(28)
    llist.push(16)
    llist.push(20)
    llist.push(1)
    
    if (llist.search(16)):
        print("The ELement is Present ")
    else: 
        print("The ELement is not Present")
```

```python
class Node:
    
    def __init__(self,data):
        self.data = data
        self.next = None
        
class LinkedList9:
    
    def __init__(self):
        self.head = None
        
    def push(self, new_data):
        
        new_node = Node(new_data)
        
        new_node.next = self.head
        
        self.head = new_node
        
    def searchRec(self, li, key):
        
        if(not li):
            return False
        
        if(li.data == key):
            return True
        
        return self.searchRec(li.next, key)
    
    
if __name__ == "__main__":
    
    llist = LinkedList9()
    
    llist.push(1)
    llist.push(2)
    llist.push(3)
    llist.push(4)
    llist.push(5)
    
    key = 4
    if llist.searchRec(llist.head, key):
        print("Element is Present")
        
    else:
        print("Element is Not Present")
```

## To find the frequency of given number in Linked List

## Detecting Loop in Linked List using Floyd's Cycle-Finding ALgorithm 

```python
class Node:
    
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList12:
    
    def __init__(self):
        self.head = None
        
    ## Function to insert node at beginning
    
    def push(self, new_data):
        
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node
        
    #Print the function
    def printList(self):
        
        temp = self.head
        while(temp != None):
            print(temp.data)
            temp = temp.next
            
    def detectLoop(self):
        
        slow_p = self.head
        fast_p = self.head
        
        while(slow_p and fast_p and fast_p.next):
            slow_p = slow_p.next
            fast_p = fast_p.next.next
            
            if slow_p == fast_p:
                print("Found Loop")
                return
            
llist = LinkedList12()
    
llist.push(20)
llist.push(4)
llist.push(15)
llist.push(10)
llist.push(13)
llist.push(14)
llist.push(16)

llist.head.next.next.next = llist.head
llist.detectLoop()
        

        
```
